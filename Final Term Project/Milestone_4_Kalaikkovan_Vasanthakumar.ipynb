```python
# creating the dataframe
df=pd.DataFrame()
```

```python
infy_json['chart']['result']
```

```python
df['timestamp']=infy_json['chart']['result'][0]['timestamp']
```

```python
df.head()
```

```python
df['low']=infy_json['chart']['result'][0]['indicators']['quote'][0]['low']
```

```python
df['volume']=infy_json['chart']['result'][0]['indicators']['quote'][0]['volume']
```

```python
df['open']=infy_json['chart']['result'][0]['indicators']['quote'][0]['open']
```

```python
df['close']=infy_json['chart']['result'][0]['indicators']['quote'][0]['close']
```

```python
df['high']=infy_json['chart']['result'][0]['indicators']['quote'][0]['high']
```

```python
df
```

### 1. Missing Value

```python
df[df['volume']==0]
```

```python
# droping rows having volume zero
df=df.drop(df[df.volume==0].index)
```

```python
df.head()
```

### 2. Format data into a more readable format

```python
from datetime import datetime
df['timestamp']=df['timestamp'].map(lambda a:datetime.fromtimestamp(a))
```

```python
df.head()
```

### 3. Identifying outliers and bad data

```python
df.drop(df[df['volume']<10000].index,inplace=True)
```

```python
df
```

### 4. Replacing Header

```python
df.rename(columns={'timestamp':'Date','low':'Low','volume':'Volume','open':'Open','high':'High'},inplace=True)
```

```python
df.head()
```

### 5. Fix casing or inconsistent values

```python
df['Date']=df['Date'].map(lambda a:str(a))
```

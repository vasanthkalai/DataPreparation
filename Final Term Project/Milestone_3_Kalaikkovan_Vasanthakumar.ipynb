```python
#importing libraries
import numpy as np
import pandas as pd
```

```python
import requests # for http requests
from bs4 import BeautifulSoup # for html parsing and scraping
import bs4 as bs
import re
import urllib
from urllib.request import Request, urlopen
```

```python
#webscraping
url = 'https://www.moneycontrol.com/stocks/marketinfo/marketcap/bse/it-services-consulting.html'

response = urllib.request.urlopen(url).read()
p = re.findall('<td((.|\s)+?)</td>', str(response)) #(pattern, string)

paragraphs = []
for x in p:
    paragraphs.append(str(x)) 
```

```python
paragraphs
```

```python
paragraphs.pop(0)
```

```python
column1=[]
column2=[]
column3=[]
column4=[]
column5=[]
column6=[]
i=1
for p in paragraphs:
    if i%6==0:
        column6.append(p)
    elif i%5==0:
        column5.append(p)
    elif i%4==0:
        column4.append(p)
    elif i%3==0:
        column3.append(p)
    elif i%2==0:
        column2.append(p)
    else:
        column1.append(p)
    i+=1
```

```python
column1=column1[0:70]
column2=column2[0:70]
column3=column3[0:70]
column4=column4[0:70]
column5=column5[0:70]
column6=column6[0:70]
d={0:column1,1:column2,2:column3,3:column4,4:column5,5:column6}
```

```python
df=pd.DataFrame(d)
```

```python
df
```

### 1. Replace Headers

```python
df.rename(columns={0:'Company Name',1:'Last Price',2:'% Change',3:'52 wk High',4:'52 wk Low',5:'Market Cap (Rs. cr)'},inplace=True)
```

```python
df.head()
```

```python
df['Company Name']=df['Company Name'].apply(lambda x: re.findall('<b>((.|\s)+?)</b>', x))
```

```python
df.head()
```

```python
df['Company Name']=df['Company Name'].apply(lambda x: x[0] if len(x)>0 else False)
```

```python
df
```

```python
df.drop(df[df['Company Name']==False].index,inplace=True)
```

```python
df
```

```python
df['Company Name']=df['Company Name'].apply(lambda x: x[0])
```

```python
df
```

```python
df['Last Price']=df['Last Price'].apply(lambda x: x.split(',')[1][0:-1])
```

```python
df['% Change']=df['% Change'].apply(lambda x: x.split('>')[1].split(',')[0][0:-1])
```

```python
df['52 wk High']=df['52 wk High'].apply(lambda x: x.split('>')[1][0:-7])
```

```python
df['52 wk Low']=df['52 wk Low'].apply(lambda x: x.split('>')[1][0:-7])
```

```python
df['Market Cap (Rs. cr)']=df['Market Cap (Rs. cr)'].apply(lambda x: x.split('>')[1][0:-7])
```

### 5.Format data into a more readable format

```python
df['Market Cap (Rs. cr)']=df['Market Cap (Rs. cr)'].apply(lambda x: x.replace(',',""))
```

```python
df['Market Cap (Rs. cr)'] = df['Market Cap (Rs. cr)'].astype(float)
```

```python
df['Market Cap (Rs. cr)']=df['Market Cap (Rs. cr)'].map(lambda x: x/1000)
```

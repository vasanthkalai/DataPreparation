## Chapter 7

#### 1. Removing duplicates

```python
print("Before removing duplicates",len(df_2015))
df_2015.drop_duplicates(subset ="Timestamp",keep = False, inplace = True)
print("After removing duplicates",len(df_2015))
```

```python
df_2015['How old are you?'].dropna()
```

```python
print("Before removing duplicates",len(df_2016))
df_2016.drop_duplicates(subset ="Timestamp",keep = False, inplace = True)
print("After removing duplicates",len(df_2016))
```

## Chapter 8

#### 3. Combine and Merge Datasets

```python
dataframes=[df_2015,df_2016]
df=pd.concat(dataframes)
```

```python
df.head()
```

### 4. Pivot the Data

```python
df.pivot(index='Timestamp',columns="How old are you?")
```

## Chapter 10

### 5. Grouping with Dicts/Series

```python
d=dict()
for i in range(1,100):
    d[i]=i

df['group']=df['How old are you?'].map(d)
grouped=df.groupby('group')

for group_name, data in grouped:
    print(group_name)
    print('---------')
    print(data)
```

### 6. Grouping with Index Levels

```python
df.set_index(['Timestamp','How old are you?'])
```

## Chapter 11

### 7. Convert between string and date time

```python
df['Timestamp'] = df['Timestamp'].dt.strftime('%Y-%m-%d')
```

```python
df.head()
```

### 8. Convert timestamps to periods and back

```python
df["Periods"]=pd.DatetimeIndex(df["Timestamp"]).to_period('M')
```
